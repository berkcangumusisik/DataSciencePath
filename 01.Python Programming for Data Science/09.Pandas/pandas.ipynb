# Koşullu Seçim (Conditional Selection)

```python
a[a["age"] > 50].head() # age sütunundaki değerler 50'den büyük olanları getirir.
```

```python
a[a["age"] > 50].count()  # age sütunu 50 den büyük olanların sayısını verir.
```

```python
a.loc[a["age"] > 50, ["age", "class"]].head()  # age sütunu 50 den büyük olanların age ve class sütunlarını seçer.
```

```python
a.loc[(a["age"] > 50) & (a["sex"] == "male"), ["age","class"]].head()  # Yaşı 50'den büyük ve cinsiyeti erkek olan age ve class koşullarını seç.
```

```python
df_new = a.loc[(a["age"] > 50) & (a["sex"] == "male") & (
        (a["embark_town"] == "Cherbourg") | (a["embark_town"] == "Southampton")), ["age", "class", "embark_town"]]
# Yaşı 50'den büyük ve cinsiyeti erkek olanları ve Cherbourg ve Southampton'da olanları seç. age, class, embark_town sütunlarını yaz.

df_new["embark_town"].value_counts()
```

## Toplulaştırma ve Gruplama (Aggregation & Grouping)
- count()
- first()
- last()
- mean()
- median()
- min()
- max()
- std()
- var()
- sum()
- pivot table

```python
# groupby() ile gruplama yapılır.
# Cinsiyete göre yaş ortalamasını ekrana yazdırır.
a.groupby("sex")["age"].mean()
```

```python
a.groupby("sex").agg({"age": "mean"})
```

```python
a.groupby("sex").agg({"age": ["mean", "sum"]})  # Cinsiyete göre yaş ortalamasını ve yaş toplamını alır.
```

```python
a.groupby("sex").agg({"age": ["mean", "sum"],
                            "survived": "mean"})
```

```python
a.groupby(["sex", "embark_town"]).agg({"age": ["mean"], "survived": "mean"})
```

```python
a.groupby(["sex", "embark_town", "class"]).agg({"age": ["mean"], "survived": "mean"})
```

```python
a.groupby(["sex", "embark_town", "class"]).agg(
    {"age": ["mean"],
     "survived": "mean",
     "sex": "count"})
```

# Merge ile Birleştirme İşlemleri


```python
df1 = pd.DataFrame({'employees': ['john', 'dennis', 'mark', 'maria'],
                    'group': ['accounting', 'engineering', 'engineering', 'hr']})

df2 = pd.DataFrame({'employees': ['mark', 'john', 'dennis', 'maria'],
                    'start_date': [2010, 2009, 2014, 2019]})

pd.merge(df1, df2)
```

```python
pd.merge(df1, df2, on="employees")
```

```python
# Amaç: Her çalışanın müdürünün bilgisine erişmek istiyoruz.
df3 = pd.merge(df1, df2)

df4 = pd.DataFrame({'group': ['accounting', 'engineering', 'hr'],
                    'manager': ['Caner', 'Mustafa', 'Berkcan']})

pd.merge(df3, df4)
```
